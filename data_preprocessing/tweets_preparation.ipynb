```python
# The original dataset found online has two columns: 'Tweet id' and 'BERT_label'. It has around 200M rows.
# The device that hydrates our tweets only manage csv file with no column header, no index. Nothing but the tweets_ID

# Let's create a DataFrame named tid with nothing but the tweets_ID 
tid = pd.read_csv("/content/drive/MyDrive/UCL/Y2/asonam_release_all_tweets.csv").copy() # Other users, do not forget to modify path
tid.drop('BERT_label', inplace=True, axis=1)
tid = tid.rename(columns={'Tweet ID': 'tweet_ID'})
tid
```

###Importing 15M tweets

```python
# Takes 1 row every of 13 rows.
# We chose this method rather than getting 10M tweets randomly because the tweet_id were harvested chronogically. 
# This will avoid having period of time with greater number of tweets than other in an unjustified way.
# This avoid inferences in later statistics for instance the evolution of hate speech over time.
tid15m = tid.sample(15000000)
tid15m
```

```python
# Remove the header and the index and save as a CSV file
tid15m.to_csv('tid15m.csv', header = False, index = False)
```
